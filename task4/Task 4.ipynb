## 2 Исследование наборов данных на портале data.worldbank.org

##### Климаттың өзгерісі жайлы деректерді таңдадым

```python
PATH_CLIMAT = "climate-change-excel-4-6-mb-.xls"
```

```python
data = pd.read_excel(PATH_CLIMAT)
```

```python
data
```

#### Қандай климаттық көрсеткіштер бар екенін қарайық

```python
series = set(data["Series name"].tolist())
series
```

### 2.4 На основе набора данных подготовьте выборку, содержащую значения показателя за все годы для трёх произвольно выбранных стран мира

```python
countries = data["Country name"].tolist()
```

#### Кездейсоқ 3 мемлекет таңдайық

```python
chosen_ctrs = np.random.choice(countries, 3)
chosen_ctrs
```

#### Таңдалған 3 мемелекеттің "Urban population" шамаларын қарастырайық

```python
chosen_data = data.loc[data["Country name"].isin(chosen_ctrs)]
chosen_data = chosen_data[chosen_data["Series name"] == "Urban population"]
chosen_data
```

### 2.5 На основе подготовленной выборки постройте график, иллюстрирующий изменение показателя со временем для трёх стран мира.

```python
def get_values(country_name, years):
    return chosen_data[chosen_data["Country name"] == country_name][years].values.tolist()[0]
```

```python
years = list(range(1990, 2011))
guatemala = get_values("Guatemala", years)
jamaica = get_values("Jamaica", years)
tunisia = get_values("Tunisia", years)
```

```python
from matplotlib.ticker import MaxNLocator

ax = plt.figure(figsize=(21, 8), dpi=80).gca()
ax.xaxis.set_major_locator(MaxNLocator(integer=True))

plt.plot(years, guatemala, label="Guatemala")
plt.plot(years, jamaica, label="Jamaica")
plt.plot(years, tunisia, label="Tunisia")
plt.legend()
plt.xlabel("Year")
plt.ylabel("Population")
plt.title("Urban population")
plt.show()
```

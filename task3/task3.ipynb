## Берілген кесте бойынша 2 векторды толтырайық. Нет - 0, есть - 1

```python
y_true = [1, 0, 0, 1, 0, 1, 1, 0, 0, 0]
y_pred = [1, 0, 0, 1, 0, 0, 1, 0, 0, 0]
```

## Вирустың таралым диаграммасын сызайық

```python
sns.countplot(y_true)
```

## Рассчитайте значения точности, чувствительности, F-критерия

```python
res = precision_recall_fscore_support(y_true, y_pred)
```

```python
print(f"Precision for class 0: {res[0][0]}")
print(f"Precision for class 1: {res[0][1]}")
print(f"Recall for class 0: {res[1][0]}")
print(f"Recall for class 1: {res[1][1]}")
print(f"F1 score for class 0: {res[2][0]}")
print(f"F1 score for class 1: {res[2][1]}")
```
